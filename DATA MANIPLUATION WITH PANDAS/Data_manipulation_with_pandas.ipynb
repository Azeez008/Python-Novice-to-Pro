### What is Pandas?

- Pandas is a Python library for working with structured data (tables, spreadsheets, CSVs, SQL query results, JSON). It gives you powerful tools to clean, explore, transform, aggregate, and visualize data
##### The two core data structures

- Series: a single labeled column (1-D).

- DataFrame: a table of rows and columns (2-D), basically a collection of Series that share the same row index.

##### Getting data in and out (I/O)

- Read: pd.read_csv(), read_excel(), read_json(), read_sql(), read_parquet().

- Write: df.to_csv(), to_excel(), to_json(), to_parquet().
