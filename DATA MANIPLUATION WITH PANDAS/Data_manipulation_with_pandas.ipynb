### What is Pandas?

- Pandas is a Python library for working with structured data (tables, spreadsheets, CSVs, SQL query results, JSON). It gives you powerful tools to clean, explore, transform, aggregate, and visualize data
##### The two core data structures

- Series: a single labeled column (1-D).

- DataFrame: a table of rows and columns (2-D), basically a collection of Series that share the same row index.

##### Getting data in and out (I/O)

- Read: pd.read_csv(), read_excel(), read_json(), read_sql(), read_parquet().

- Write: df.to_csv(), to_excel(), to_json(), to_parquet().

```python
 #keep the latest year per city (if multiple years exist)
latest = (df_dummy.sort_values("year")
                   .groupby("city", as_index=False)
                   .tail(1)
                   .sort_values("city"))

cities = latest["city"].tolist()
sales = latest["sales"].tolist()
costs  = latest["cost"].tolist()

x = range(len(cities))
width = 0.35

plt.figure(figsize=(8,4))
plt.bar([i - width/2 for i in x], sales, width, label="Sales")
plt.bar([i + width/2 for i in x], costs, width, label="Cost")
plt.xticks(x, cities)
plt.ylabel("Amount")
plt.title("Sales vs Cost by City (Latest Year per City)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Seaborn
#Seaborn: Line plot of Sales over Time by City
#This shows sales trend by year for each city (useful when you add more years).


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# (reusing df_dummy from above)

plt.figure(figsize=(8,4))
sns.lineplot(
    data=df_dummy.sort_values("year"),
    x="year", y="sales", hue="city", marker="o"
)
plt.title("Sales Trend by City")
plt.ylabel("Sales")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```
