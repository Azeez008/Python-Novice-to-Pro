```python
#Create a function sq_cube which returns a list of lists containing both the square and the cube of only the even numbers in a given list. The function must also be able to round the numbers to the nearest whole number
```

```python
e.g [1 ,2, 4, 5.6]
[[4,8], [16,64], [36,216]]
```

```python
def sq_cube(numbers):
    num_var = []
    for t in numbers:
        num_var.append(int(round(t,0)))
    even_var = []    
    for var in num_var:
        if var % 2==0:
            even_var.append(var)
    s=[]
    c =[]        
    for nums in even_var:
        s.append(nums**2)
        c.append(nums**3)
    for i in s,c:
        combined_list = list(list(zip(s,c)))    
        
    final_list = [list(i) for i in combined_list] 
   
   
    
    return final_list
```

```python
def sq_cube(numbers):
    rounded_up=[round (num) for num in numbers]
    even_num=[]
    for num in rounded_up:
        if num % 2==0:
            even_num.append(num)
    final_num=[[(num)**2, (num)**3] for num in even_num]

    return final_num
```

Challenge
Proteins are the building blocks of life. The process starts when another protein (a polymerase) reads your genetic code and performs transcription, the code (RNA) for making a protein. Ribosomes will then attach to this RNA code to begin making specific proteins. Your genetic code is made up of pairs of nucleotides. Once transcribed into the code, these nucleotides contain the following nucleotides: U,G,A,C. UGA, UAA, and UAG are stop codons, and AUG is a start codon so that the protein, ribosome, can identify where to stop and start the protein formation. A genetics lab has tasked you to design a function that recognises the length of the RNA code (excluding the start and stop codons). The function must also determine if the code is actually a valid piece of code (must contain AUG as a start codon and UGA, UAA, or UAG at the end).

Note: RNA starts with a start codon AUG and ends with UGA, UAA, or UAG
The function must return "Not readable RNA code" if the above conditions are not met.

```python
def rna_code_length(c):
    if c.startswith("AUG") and c.endswith("UGA"):
        return len(c) - 6
    elif c.startswith("AUG") and c.endswith("UAA"):
        return len(c) - 6
    elif c.startswith("AUG") and c.endswith("UAG"):
        return len(c) - 6
    else:
        return "Not readable RNA code"
```

```python
rna_code_length('AUGAGGCACCUUCUGCUCCUUAC')
```

```python
rna_code_length('AUGUAGCAUAA')
```

```python
rna_code_length('AUGUAGGCACAUUUAUGCUCCUGA')
```

```python
def rna_length(RNA):
    if RNA[0:3]== 'AUG' and RNA[-3:] in ('UGA','UAA','UAG'):
        RNA = RNA[3:]
        RNA = RNA[0 :-3]
        result = len(RNA)
        return result
    else :
        return "Not readable RNA code"
```
